```python
# Compute mortality on training dataset for old state
state_num = 750
action_num = 25

state_action_m = np.zeros((state_num, action_num, 2), dtype=np.float32)
state_m = np.zeros((state_num, 2), dtype=np.float32)


# Sum up
#train_df["wt_mort"]=0
for hosp_id in train_df.hosp_id.unique().tolist():
  #print("Hosp_id:", hosp_id)
  subset_df = train_df[train_df.hosp_id == hosp_id]
  
  n_state = len(subset_df)
  if n_state == 0:
    continue  
  weight = 1.0 / n_state
  
  for idx in subset_df.index:
    state = int(subset_df.loc[idx, "state"] - 1)
    action = int(subset_df.loc[idx, "action"] - 1)
    outcome = float(subset_df.loc[idx, "dead_90"])    
        
    state_action_m[state][action][0] += outcome * weight
    state_action_m[state][action][1] += weight
    
    state_m[state][0] += outcome * weight
    state_m[state][1] += weight
  
# Normalize
for state in range(state_num):
  weight = state_m[state][1]
  if weight > 0.0:
    state_m[state][0] /= weight  
  
  for action in range(action_num):
    weight = state_action_m[state][action][1]
    if weight > 0.0:
      state_action_m[state][action][0] /= weight
    else:
      state_action_m[state][action][0] = state_m[state][0]
```

```python

import numpy as gfg
  
# reshaping the array from 3D
# matrice to 2D matrice.
arr_reshaped = state_action_m.reshape(state_action_m.shape[0], -1)
  
# saving reshaped array to file.
gfg.savetxt("data_noshuffle\state_action_m.txt", arr_reshaped)
  
# retrieving data from file.
#loaded_arr = gfg.loadtxt("geekfile.txt")
```

```python
# Compute mortality on training dataset for new state
state_num = 750
action_num = 25

state_action_m_new = np.zeros((state_num, action_num, 2), dtype=np.float32)
state_m_new = np.zeros((state_num, 2), dtype=np.float32)


# Sum up
#train_df["wt_mort_new"]=0
for hosp_id in train_df.hosp_id.unique().tolist():
  #print("Hosp_id:", hosp_id)
  subset_df = train_df[train_df.hosp_id == hosp_id]
  
  n_state = len(subset_df)
  if n_state == 0:
    continue  
  weight = 1.0 / n_state
  
  for idx in subset_df.index:
    state = int(subset_df.loc[idx, "new_state"] - 1)
    action = int(subset_df.loc[idx, "action"] - 1)
    outcome = float(subset_df.loc[idx, "dead_90"])    
        
    state_action_m_new[state][action][0] += outcome * weight
    state_action_m_new[state][action][1] += weight
    
    state_m_new[state][0] += outcome * weight
    state_m_new[state][1] += weight
  
# Normalize
for state in range(state_num):
  weight = state_m_new[state][1]
  if weight > 0.0:
    state_m_new[state][0] /= weight  
  
  for action in range(action_num):
    weight = state_action_m_new[state][action][1]
    if weight > 0.0:
      state_action_m_new[state][action][0] /= weight
    else:
      state_action_m_new[state][action][0] = state_m_new[state][0]
```

```python
from random import sample
def comput_mort(df,state_input,policy,state_action_m,n,nboot):
  unique_id=(test_df.hosp_id).unique().tolist()
  mort = []
  
  nboot =nboot
  for i in range(nboot):
    sample_id = sample(unique_id,n)
    test_df1=test_df[test_df.hosp_id.isin (sample_id)]
    test_df1 = test_df1.reset_index(drop=True)
    tot_m =0.0
    add_cnt = 0
    for test_id in test_df1.index:
      state = int(test_df1.loc[test_id, state_input] - 1)

      action = policy.loc[state, "action"] - 1
      
      m = state_action_m[state][action][0]
      tot_m += m
      add_cnt += 1
    mort.append(tot_m/add_cnt)
 
  return mort
```
