```python
import pandas as pd
import numpy as np
import seaborn as sb
from matplotlib.pyplot import figure

import matplotlib.pyplot as plt
```

```python
wis_df = pd.read_csv("data_noshuffle\wis_df_lastversion.csv")

wis_df = wis_df.drop(columns={'Unnamed: 0'})
#wis_df = abs(wis_df)
wis_intm1 = pd.read_csv("data_noshuffle\wis_df_intm1.csv")
wis_intm1 = wis_intm1.drop(columns={'Unnamed: 0'})
#wis_intm1 = abs(wis_intm1)
wis_intm2 = pd.read_csv("data_noshuffle\wis_df_intm2.csv")
wis_intm2 = wis_intm2.drop(columns={'Unnamed: 0'})
wis_intm3 = pd.read_csv("data_noshuffle\wis_df_intm3.csv")
wis_intm3 = wis_intm3.drop(columns={'Unnamed: 0'})

#wis_intm2 = abs(wis_intm2)
wis_vaso1 = pd.read_csv("data_noshuffle\wis_df_vaso1.csv")
wis_vaso1 = wis_vaso1.drop(columns={'Unnamed: 0'})
#wis_vaso1 = abs(wis_vaso1)
wis_vaso2 = pd.read_csv("data_noshuffle\wis_df_vaso2.csv")
wis_vaso2 = wis_vaso2.drop(columns={'Unnamed: 0'})
#wis_vaso2 = abs(wis_vaso2)
wis_vaso3 = pd.read_csv("data_noshuffle\wis_df_vaso3.csv")
wis_vaso3 = wis_vaso3.drop(columns={'Unnamed: 0'})
#wis_vaso2 = abs(wis_vaso2)

is_mort = pd.read_csv("data_noshuffle\is_mort.csv")
```

```python
def aggr_table(df):
  s1 = df.mean(axis=0)
  s2 = df.std(axis=0)
  return pd.concat([s1, s2], axis=1)
```

```python
wis_df_agg = aggr_table(wis_df)
wis_intm1_agg = aggr_table(wis_intm1)
wis_intm2_agg = aggr_table(wis_intm2)
wis_intm3_agg = aggr_table(wis_intm3)
wis_vaso1_agg = aggr_table(wis_vaso1)
wis_vaso2_agg = aggr_table(wis_vaso2)
wis_vaso3_agg = aggr_table(wis_vaso3)
```

```python
is_mort_agg = aggr_table(is_mort)
```
